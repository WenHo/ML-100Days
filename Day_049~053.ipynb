```python
# Import 需要的套件
import os
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
# 設定 data_path
dir_data = './data/ml100marathon/'
f_app_train = os.path.join(dir_data, 'train_data.csv')
f_app_test = os.path.join(dir_data, 'test_features.csv')
f_app_sample = os.path.join(dir_data, 'sample_submission.csv')

train = pd.read_csv(f_app_train)
test = pd.read_csv(f_app_test)
sample = pd.read_csv(f_app_sample)

train_Y = np.log1p(sample['poi'])
```

```python
test = test.drop([20])
train_Y = train_Y.drop([20])
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(test.dtypes, test.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
test = test[num_features]
train_num = train_Y.shape[0]
# 空值補0
test_0 = test.fillna(0)
test_0.head()
```

```python
# 空值補中位數
test_0 = test.fillna(0)
train_X = test_0
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
