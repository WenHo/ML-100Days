```python
# Import 需要的套件
import os
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# 設定 data_path
dir_data = './data/ml100marathon/'
f_app_train = os.path.join(dir_data, 'train_data.csv')
f_app_test = os.path.join(dir_data, 'test_features.csv')
f_app_sample = os.path.join(dir_data, 'sample_submission.csv')

train = pd.read_csv(f_app_train)
test = pd.read_csv(f_app_test)
sample = pd.read_csv(f_app_sample)
```

```python
test.head()
```

```python
#填補遺失值
test['bonus'] = test['bonus'].fillna(test['bonus'].median())
test['deferral_payments'] = test['deferral_payments'].fillna(test['deferral_payments'].median())
test['deferred_income'] = test['deferred_income'].fillna(test['deferred_income'].median())
test['director_fees'] = test['director_fees'].fillna(test['director_fees'].median())
test['exercised_stock_options'] = test['exercised_stock_options'].fillna(test['exercised_stock_options'].median())
test['expenses'] = test['expenses'].fillna(test['expenses'].median())
test['from_messages'] = test['from_messages'].fillna(test['from_messages'].median())
test['from_poi_to_this_person'] = test['from_poi_to_this_person'].fillna(test['from_poi_to_this_person'].median())
test['from_this_person_to_poi'] = test['from_this_person_to_poi'].fillna(test['from_this_person_to_poi'].median())
test['loan_advances'] = test['loan_advances'].fillna(test['loan_advances'].median())
test['long_term_incentive'] = test['long_term_incentive'].fillna(test['long_term_incentive'].median())
test['other'] = test['other'].fillna(test['other'].median())
test['restricted_stock'] = test['restricted_stock'].fillna(test['restricted_stock'].median())
test['restricted_stock_deferred'] = test['restricted_stock_deferred'].fillna(test['restricted_stock_deferred'].median())
test['salary'] = test['salary'].fillna(test['salary'].median())
test['shared_receipt_with_poi'] = test['shared_receipt_with_poi'].fillna(test['shared_receipt_with_poi'].median())
test['to_messages'] = test['to_messages'].fillna(test['to_messages'].median())
test['total_payments'] = test['total_payments'].fillna(test['total_payments'].median())
test['total_stock_value'] = test['total_stock_value'].fillna(test['total_stock_value'].median())

#Total outliter
test.drop([20])
```

```python
from sklearn.preprocessing import MinMaxScaler, Imputer
# Scale each feature to 0-1
scaler = MinMaxScaler(feature_range = (0, 1))


# Repeat with the scaler
scaler.fit(test['bonus','deferral_payments','deferred_income','director_fees','exercised_stock_options','expenses','from_messages','from_poi_to_this_person','from_this_person_to_poi','loan_advances','long_term_incentive','other','restricted_stock','restricted_stock_deferred','salary','shared_receipt_with_poi','to_messages','total_payments','total_stock_value'])
test = scaler.transform(test['bonus','deferral_payments','deferred_income','director_fees','exercised_stock_options','expenses','from_messages','from_poi_to_this_person','from_this_person_to_poi','loan_advances','long_term_incentive','other','restricted_stock','restricted_stock_deferred','salary','shared_receipt_with_poi','to_messages','total_payments','total_stock_value'])
```
